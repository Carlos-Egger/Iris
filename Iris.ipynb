```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
pd.__version__, np.__version__, tf.__version__, sklearn.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv('iris.csv')
base
```

```python
X = base.iloc[:, 0:4].values
X
```

```python
y = base.iloc[:, 4].values
y
```

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
```

```python
y
```

```python
y = np_utils.to_categorical(y)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),
    tf.keras.layers.Dense(units=4, activation = 'relu'),
    tf.keras.layers.Dense(units=4, activation = 'relu'),
    tf.keras.layers.Dense(units=3, activation = 'softmax'),

])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
rede_neural.fit(X_train, y_train, batch_size = 10, epochs = 1000)
```

```python
rede_neural.evaluate(X_test, y_test)
```

```python
previsoes = rede_neural.predict(X_test)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
```

```python
print(previsoes)
```

```python
y_test2 = [np.argmax(t) for t in y_test]
print(y_test2)
```

```python
previsoes2 = [np.argmax(t) for t in previsoes]
print(previsoes2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test2, previsoes2)
```

```python
confusion_matrix(y_test2, previsoes2)
```
